```python
from tensorflow.keras.models import load_model
import numpy as np
import PIL
from PIL import Image
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
model = load_model("VGG16trainedmodel.h5")
```

```python
im1 = Image.open('D://FYP Project//Dataset//Test//Real//real_1078.jpg')
im2=im1.resize((224,224))
p1 = np.array(im2)
p1=p1/255
plt.imshow(p1)
p1 = np.expand_dims(p1, axis=0)
p1.shape
```

```python
p=np.argmax(model.predict(p1),axis=1)
print(p)
if p==[0]:
    print('Fake Image')
else:
    print('Real Image')
```

```python
ig = ImageDataGenerator(rescale=1./255.)
```

```python
test_flow = ig.flow_from_directory(
    "D://FYP Project//Dataset//Test",
    target_size=(224, 224),
    batch_size=64,
    shuffle = False,
    class_mode='categorical'
)
```

```python
y_pred = model.predict(test_flow)
```

```python
for i in y_pred:
    print(i)
```

```python
y_test = test_flow.classes
```

```python
for i in y_test:
    print(i)
```

```python
y_pred_labels = np.argmax(y_pred, axis=1)
```

```python
confusion_matrix=confusion_matrix(y_test, y_pred_labels)
confusion_matrix
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

```python
print("ROC AUC Score:", metrics.roc_auc_score(y_test, y_pred_labels))
print("AP Score:", metrics.average_precision_score(y_test, y_pred_labels))
print()
print(metrics.classification_report(y_test, y_pred_labels))
```

```python
_, accu = model.evaluate(test_flow)
print('Final Test Acccuracy = {:.3f}'.format(accu*100))
```
