```python
from tkinter import *
from PIL import ImageTk, Image
from ipynb.fs.full.Voice_setup import call_voice
from ipynb.fs.full.About_GUI import about
from ipynb.fs.full.Deepfake_GUI import deepfaketest

class main:
    def __init__(self, root):
        self.root = root
        root.geometry("1400x600")
        root.maxsize(1400,600)
        root.minsize(1400,600)
        
        root.title("Anti Deepfake Examiner")
        
        
        Header_frame = Frame(root, bg="medium spring green", height=150)
        
        Header_block1 = Frame(Header_frame, bg="medium spring green", height=68)
        title_label = Label(Header_block1, text="Anti Deepfake Examiner", bg="medium spring green",
                            font=('Garuda', 30, 'bold', 'italic'))
        tagline_label = Label(Header_block1, text="(\"Guarding Truth: Unmasking Deception with Deepfake Vigilance\")",
                             bg="medium spring green", fg="navy",
                            font=('Garuda', 15, 'bold', 'italic'))
        
        
        Center_frame = Frame(root, bg="orchid", height=450)
        
        Center_block1 = Frame(Center_frame, bg="orchid", width=400, height=500)
        self.logo_pic = Image.open("logo.png")
        self.mypic = ImageTk.PhotoImage(self.logo_pic)
        logo_label = Label(Center_block1, image=self.mypic, bg="orchid")
        
        Center_block2 = Frame(Center_frame, bg="orchid", width=600 ,height=500)
        App_about_label = Label(Center_block2, justify='center', font=('Garuda', 12, 'italic'), bg='DodgerBlue3', text=
        """
        In an era where visual content dominates the digital landscape,    
    the rise of deepfake technology has introduced unprecedented    
    challenges to the authenticity of images. Our cutting-edge    
    application, Unmasking Reality emerges as a powerful solution    
    to combat the deceptive world of deepfake images.    
    \n\n
    the User Friendly Interface, emphasizing its simplicity and    
    accessibility. Users should be able to easily upload images for    
    analysis and receive clear results.    
        """)
        
        
        Center_block3 = Frame(Center_frame, bg="orchid", width=700, height=500)
        Deepfake_button = Button(Center_block3, text="Test Deepfake", bg="DodgerBlue3", font=('Garuda',12,'bold'), width=20, 
                            height=1, fg="mint cream", activebackground="mint cream", activeforeground="DodgerBlue3",
                            command= lambda : deepfaketest(root))
        About_button = Button(Center_block3, text="About Coder Tech", bg="DodgerBlue3", font=('Garuda',12,'bold'), width=20, 
                            height=1, fg="mint cream", activebackground="mint cream", activeforeground="DodgerBlue3",
                            command= lambda : about(root))
        
        
        Footer_frame = Frame(root, bg="medium spring green", height=50)
        copytight_label = Label(Footer_frame, text="Â© CODER TECH 2023, All Rights Reserved", bg="medium spring green",
                                font=('Garuda',12,'bold'))
        
        
        Header_frame.pack(side=TOP, fill=BOTH)
        Header_block1.pack(side=TOP, fill=BOTH, pady=15)
        title_label.pack(pady=(0,3))
        tagline_label.pack()
        
        
        Center_frame.pack(side=TOP, fill=BOTH)
        
        Center_block1.pack(side=LEFT, fill=BOTH, expand=1)
        logo_label.pack(pady=25, padx=50)
        
        
        Center_block2.pack(side=LEFT, fill=BOTH, expand=1)
        App_about_label.pack(padx=50, pady=(100,0))
        
        
        Center_block3.pack(side=LEFT, fill=BOTH, expand=1)
        Deepfake_button.pack(padx=50, pady=(180,0))
        About_button.pack(padx=50, pady=(20,0))

        Footer_frame.pack(side=BOTTOM, fill=BOTH)
        copytight_label.pack(pady=5)
        
        call_voice.wishme()
        
root = Tk()
obj = main(root)
mainloop()
```
