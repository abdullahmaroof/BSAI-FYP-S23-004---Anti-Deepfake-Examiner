```python
from tkinter import *
from PIL import ImageTk, Image
from tkinter import filedialog
from ipynb.fs.full.Voice_setup import *
import numpy as np
from tensorflow.keras.models import load_model
import tkinter.messagebox as mg

def deepfaketest(root):
    root.destroy()
    class deepfakeportal:       
        def __init__(self, root):
            self.root = root
            root.geometry("1400x698")
            root.maxsize(1400,698)
            root.minsize(1400,698)

            root.title("Anti Deepfake Examiner/Detection Portal")


            Header_frame = Frame(root, bg="medium spring green", height=150)

            Header_block1 = Frame(Header_frame, bg="medium spring green", height=68)
            title_label = Label(Header_block1, text="Anti Deepfake Examiner", bg="medium spring green",
                                font=('Garuda', 30, 'bold', 'italic'))
            tagline_label = Label(Header_block1, text="(\"Guarding Truth: Unmasking Deception with Deepfake Vigilance\")",
                                 bg="medium spring green", fg="navy",
                                font=('Garuda', 15, 'bold', 'italic'))


            Center_frame = Frame(root, bg="orchid", height=550)

            centerfame_top = Frame(Center_frame, bg="orchid", height=100)
            Sectionheading_label = Label(centerfame_top, text="Deepfake Test Portal", font=('Garuda', 20, 'italic', 'bold'), 
                                           bg="DodgerBlue3", width=30, height=2, fg='white')

            centerfame_bottom = Frame(Center_frame, bg="orchid", height=450)

            Center_block1 = Frame(centerfame_bottom, bg="orchid", width=400, height=450)
            self.logo_pic = Image.open("logo.png")
            self.mypic = ImageTk.PhotoImage(self.logo_pic)
            logo_label = Label(Center_block1, image=self.mypic, bg="orchid")


            Center_block2 = Frame(centerfame_bottom, bg="orchid", width=300 ,height=450)
            Uploadimage_button = Button(Center_block2, text="Upload Image", bg="DodgerBlue3", font=('Garuda',12,'bold'), width=20, 
                                height=1, fg="mint cream", activebackground="mint cream", activeforeground="DodgerBlue3")
            Test_button = Button(Center_block2, text="Test Image", bg="DodgerBlue3", font=('Garuda',12,'bold'), width=20, 
                                height=1, fg="mint cream", activebackground="mint cream", activeforeground="DodgerBlue3",)


            Center_borderblock1 = Frame(centerfame_bottom, bg="black", width=2, height=450)

            Center_block3 = Frame(centerfame_bottom, bg="orchid", width=700, height=450)
            Imagepath_Entry = Entry(Center_block3, font=('Garuda',10), width=70)
            UploadImage_label = Label(Center_block3, bg="orchid")




            Footer_frame = Frame(root, bg="medium spring green", height=50)
            copytight_label = Label(Footer_frame, text="Â© CODER TECH 2023, All Rights Reserved", bg="medium spring green",
                                    font=('Garuda',12,'bold'))

            def selectPic():
                global img
                filename = filedialog.askopenfilename(initialdir='C://Users//Abdullah Maroof//BSAI-FYP-S23-004//Dataset//Test', title="Select Image To Test",
                                       filetypes=(("jpg images","*.jpg"),("png images","*.png")))
                img = Image.open(filename)
                img = img.resize((400,300), Image.Resampling.LANCZOS)
                img = ImageTk.PhotoImage(img)
                Imagepath_Entry.insert(0, filename)
                UploadImage_label['image'] = img
                file = open("imagelibk.txt","w")
                file.write(filename)
                file.close()

            def clear_entry():
                Imagepath_Entry.delete(0, 'end')


            def test_deepfake():
                Imagepath_Entry.delete(0, 'end')
                model = load_model("Resnet_50trainedmodel.h5")
                file = open("imagelibk.txt","r")
                link = file.read()
                file.close()
                im1 = Image.open(link)
                im2=im1.resize((128,128))
                p1 = np.array(im2)
                p1=p1/255
                p1 = np.expand_dims(p1, axis=0)
                p1.shape
                p=np.argmax(model.predict(p1),axis=1)
                print(p)
                file = open("Resultsfile.txt","a")
                file.write(link)
                if p==[0]:
                    print('Fake Image')
                    mg.showinfo("Test Successfull", "This Image is fake")
                    voice_fakeimage.wishme()
                    file.write('\nFake Image\n')
                else:
                    print('Real Image')
                    mg.showinfo("Test Successfull", "This Image is Real")
                    voice_realimage.wishme()
                    file.write('\nReal Image\n')
                file.close()


            Uploadimage_button['command'] = lambda : [clear_entry(), selectPic()]
            Test_button['command'] = test_deepfake

            Header_frame.pack(side=TOP, fill=BOTH)
            Header_block1.pack(side=TOP, fill=BOTH, pady=15)
            title_label.pack(pady=(0,3))
            tagline_label.pack()


            Center_frame.pack(side=TOP, fill=BOTH, pady=(2,0))

            centerfame_top.pack(side=TOP, fill=BOTH, expand=1)
            Sectionheading_label.pack(pady=10)


            centerfame_bottom.pack(side=TOP, fill=BOTH, expand=1, pady=(2,0))

            Center_block1.pack(side=LEFT, fill=BOTH, expand=1)
            logo_label.pack(pady=25, padx=50)


            Center_block2.pack(side=LEFT, fill=BOTH, expand=1)
            Uploadimage_button.pack(padx=50, pady=(180,0))
            Test_button.pack(padx=50, pady=(20,0))

            Center_borderblock1.pack(side=LEFT, fill=BOTH,pady=(0,5))

            Center_block3.pack(side=LEFT, fill=BOTH, expand=1)
            Imagepath_Entry.pack(side=TOP, padx=20, pady=(20,40))
            UploadImage_label.pack(side=TOP, padx=20, pady=(0,20))

            Footer_frame.pack(side=BOTTOM, fill=BOTH)
            copytight_label.pack(pady=5)

            voice_deepfakeportal.wishme()


    root = Tk()
    obj = deepfakeportal(root)
    mainloop()   
   
        
```
